```python
# Criar Coluna "Mês" a partir da coluna "Data da colheita"
dataset["Mês"] = dataset["Data da colheita"].dt.month
```

```python
mes = dataset.groupby("Mês").agg(Número_Amostras_Colhidas=("Mês","count")).reset_index()
mes
```

```python
def mes_nome(mes):
    if mes == 1:
        return "Jan"
    elif mes == 2:
        return "Fev"
    elif mes == 3:
        return "Mar"
    elif mes == 4:
        return "Abr"
    elif mes == 5:
        return "Mai"
    elif mes == 6:
        return "Jun"
    elif mes == 7:
        return "Jul"
    elif mes == 8:
        return "Ag"
    elif mes == 9:
        return "Set"
    elif mes == 10:
        return "Out"
    elif mes == 11:
        return "Nov"
    elif mes == 12:
        return "Dez"
```

```python
mes["Mes_nome"] = mes["Mês"].apply(mes_nome)
```

```python
mes
```

```python
# Create figure and axes
fig, ax = plt.subplots(figsize=(12, 6))

# Set background color
fig.set_facecolor('#EAEAF2')

# Set font style
sns.set(font_scale=1.2)

# Set blue color palette
green_palette = sns.color_palette("Greens_r")

# Set color for plot
color = green_palette[2]

# Create line plot
sns.lineplot(x=mes["Mes_nome"], y=mes["Número_Amostras_Colhidas"], linewidth=1, color=color, marker="o")

# Add values as annotations
for x, y in zip(mes["Mes_nome"], mes["Número_Amostras_Colhidas"]):
    plt.annotate(text=str(y), xy=(x, y), xytext=(5, -5), textcoords='offset points', ha='center', va='top', fontsize=13)

# Add labels and title
plt.title("Número de amostras colhidas por mês", fontsize=18, fontweight="bold")
plt.xlabel("Mês", fontsize=16)
plt.ylabel("Número de amostras", fontsize=16)

# Customize ticks and grid
ax.set_xticklabels(mes["Mes_nome"], rotation=45, ha='right', fontsize=16)
ax.set_yticklabels(ax.get_yticks(), fontsize=16)
plt.grid(axis='y', alpha=0.4)

# Remove top and right spines
sns.despine(top=True, right=True)

# Show plot
plt.show()
```

```python
def plot_nc(data,title):
    # Create figure and axes
    fig, ax = plt.subplots(figsize=(12, 5))

    # Set background color
    fig.set_facecolor('#EAEAF2')

    # Set font style
    sns.set(font_scale=1.2)

    # Set blue color palette
    green_palette = sns.color_palette("Greens_r")

    # Set color for plot
    color = green_palette[2]

    # Create line plot
    sns.lineplot(x=data["index"], y=data["%NC"], linewidth=1, color=color, marker="o")
    
    for x, y in zip(data["index"], data["%NC"]):
        plt.annotate(text=f"{y:.2f}%", xy=(x, y), xytext=(5, -5), textcoords='offset points', ha='center', va='top', fontsize=11)

    # Add labels and title
    plt.title(title, fontsize=18,fontweight="bold")
    plt.xlabel("Grupo", fontsize=16,rotation=0)
    plt.ylabel("% NC", fontsize=14)

    # Customize ticks and grid
    ax.set_xticklabels(data["index"], rotation=50, ha='right', fontsize=12)
    ax.set_yticklabels(ax.get_yticks(), fontsize=12)
    plt.grid(axis='y', alpha=0.4)
```

```python
plot_nc(fraude,"% NC/Fraude 2022")
```
